```python
import pandas as pd
import re
```

```python
train_1 = pd.read_csv("web_traffic_forecasting/resources/train_1.csv.zip")
train_2 = pd.read_csv("web_traffic_forecasting/resources/train_2.csv.zip")

key_1 = pd.read_csv("web_traffic_forecasting/resources/key_1.csv.zip")
key_2 = pd.read_csv("web_traffic_forecasting/resources/key_2.csv.zip")
```

```python
def get_language(page):
    res = re.search('[a-z][a-z].wikipedia.org',page)
    if res:
        return res[0][0:2]
    return 'na'
```

```python
train_1.isnull().sum()
```

```python
train_2.isnull().sum()
```

```python
train_1.fillna(0, inplace=True)
train_2.fillna(0, inplace=True)
```

```python
train_1['lang'] = train_1.Page.map(get_language)
train_2['lang'] = train_2.Page.map(get_language)
```
